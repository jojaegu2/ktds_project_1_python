```python
#필요한 정보만 추출하기
df2=df.filter(['상호명','위도','경도'])
df2
```

```python
#문제가되는 결측치를 제거
df2_nan=df2.dropna()
df2_nan
```

```python
import requests 
import json 
from tqdm import tqdm 
import pandas as pd
import folium 
from folium.plugins import MarkerCluster, MiniMap

map_hs = folium.Map((37.260219,127.479824), zoom_start = 13)
mc = MarkerCluster()
#행 우선 접근 방식으로 값 추출하기
names=list(df2['상호명'])
lat=list(df2['위도'])
lng=list(df2['경도'])

for i in tqdm(range(len(names))):
    mc.add_child(folium.Marker(location = [lat[i], lng[i]], popup=names[i]))
    map_hs.add_child(mc)
```

```python
map_hs.save("./store_data.html")
```
